```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize']= (20.0, 10.0)


data = pd.read_csv('/home/kamzon/2021/2021/IA/Machine_learning_practice_01/ML_Prac_Iris_dataset_1.0/archive/headbrain.csv')
print(data.shape)
data.head()
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
# equation y = m*x + c

# mean Xm , Ym
X_mean = np.mean(X)
Y_mean = np.mean(Y)
s1=0
s2=0
# m= ?
for i in range(len(X)):
    s1+=(X[i]-X_mean)*(Y[i]-Y_mean)
    s2+= (X[i]-X_mean)**2
    
m=s1/s2

# c=?

c=Y_mean - X_mean*m

# y = m*x + c

print("y= ",m,"*x + ", c)
```

```python
# Ploting regression line

X_max = np.max(X) + 100
X_min = np.min(X) - 100

# calculating line values x and y

x = np.linspace(X_min, X_max, 1000)
y = c + m*x

plt.plot(x, y, color='#58b970', label='Regression Line')

plt.scatter(X, Y, c='#ef5423', label= 'scatter plot')

plt.xlabel('head size in cm3')
plt.ylabel('head weight in grams')
plt.legend()
plt.show()
```

```python
# R square value

r1=0
r2=0

for i in range(len(X)):
    Y_pred= c + m*X[i]
    r1 += (Y_pred - Y_mean)**2
    r2 += (Y[i] - Y_mean)**2
    
R2=r1/r2

print(R2)
```

```python
# # # using Sklearn librarie

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X= X.reshape((-1,1))

reg= LinearRegression()

reg= reg.fit(X,Y)

Y_pr= reg.predict(X)

r2_score = reg.score(X,Y)

print(r2_score)
```
